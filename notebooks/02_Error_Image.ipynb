```python
from PIL import Image
from concurrent.futures import ThreadPoolExecutor, as_completed
import os, glob

files = sorted(glob.glob('../data/nvidia_warehouse_dataset/train/images/*.png'))

def check_image(f):
    try:
        Image.open(f).convert('RGB')
        return None
    except:
        return os.path.basename(f)

bad = []
with ThreadPoolExecutor(max_workers=16) as pool:
    futures = {pool.submit(check_image, f): f for f in files}
    for i, future in enumerate(as_completed(futures), 1):
        result = future.result()
        if result:
            bad.append(result)
        if i % 5000 == 0:
            print(f"  Checked {i}/{len(files)}...")

print(f'\n{len(bad)} corrupted images:')
for b in sorted(bad):
    print(f'  {b}')
```
